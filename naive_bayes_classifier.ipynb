```python
import pandas as pd
import numpy as np
```

```python
outdoor_game = pd.read_csv("tennis.csv")
outdoor_game.head()

training_set = outdoor_game[0:12].values
unique_values = np.unique(outdoor_game["play"])
```

```python
def compute_conditional_probability(attribute1, attribute2): # compute P(A|B)
    numerator = sum(1 for element in training_set if attribute1 in element and attribute2 in element)
    denominator = sum(1 for element in training_set if attribute2 in element)
    return (numerator, denominator)

def compute_probability(vector, value): # compute P(Y = value | X = vector)
    print(value)
    numerators = [0 for i in range(len(vector))]
    denominators = [0 for i in range(len(vector))]

    for i in range(len(vector)):
        attribute = vector[i]
        res = compute_conditional_probability(attribute, value)
        numerators[i] += res[0]
        if numerators[i] != 0:
            denominators[i] += res[1]
        else:
            smoothing_numerator = 1
            smoothing_denominator = sum(1 for element in training_set if attribute in element)
            numerators = [element + smoothing_numerator for element in numerators]
            denominators = [element + smoothing_denominator for element in denominators]
            
    print(numerators)
    print(denominators)
    res = 1
    for i in range(len(numerators)):
        res *= numerators[i] / denominators[i]
    res *= sum(1 for element in training_set if value in element) / training_set.shape[0] # P(value)
    print(res)
    return res 

# Classify vector from outdoor game dataset using naive bayesian method.
def naive_bayes(vector):
    probabilities = np.array([compute_probability(vector, value) for value in unique_values])
    print("result: ", end=" ")
    print(unique_values[np.argmax(probabilities)])
```

```python
vector = ['sunny', 'hot', 'normal', True]
naive_bayes(vector)
```
